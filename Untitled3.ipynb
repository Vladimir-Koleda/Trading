```python
#   
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 1.   ,     
stocks = ['AAPL', 'MSFT', 'GOOGL']

# 2.   ,     
start_date = '2023-01-01'
end_date = '2023-12-31'

# 3.       yfinance
stock_data = yf.download(stocks, start=start_date, end=end_date)['Adj Close']

# 4.         
monthly_returns = stock_data.resample('M').ffill().pct_change()

# 5.      
plt.figure(figsize=(10, 6))
for stock in stocks:
    plt.plot(monthly_returns.index, monthly_returns[stock], label=stock)

# 6.      
plt.title('     ')
plt.xlabel('')
plt.ylabel('  ')

# 7.  
plt.legend()

# 8.  
plt.show()
```
