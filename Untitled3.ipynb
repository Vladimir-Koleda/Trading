```python
# Импорт необходимых библиотек
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 1. Задаем список акций, для которых хотим получить котировки
stocks = ['AAPL', 'MSFT', 'GOOGL']

# 2. Задаем временной период, за который хотим получить котировки
start_date = '2023-01-01'
end_date = '2023-12-31'

# 3. Загружаем данные о котировках с помощью yfinance
stock_data = yf.download(stocks, start=start_date, end=end_date)['Adj Close']

# 4. Создаем таблицу для процентного изменения цен акций по месяцам
monthly_returns = stock_data.resample('M').ffill().pct_change()

# 5. Построение графика процентного изменения цен акций
plt.figure(figsize=(10, 6))
for stock in stocks:
    plt.plot(monthly_returns.index, monthly_returns[stock], label=stock)

# 6. Добавляем заголовок и метки к осям
plt.title('Процентное изменение цен акций по месяцам')
plt.xlabel('Дата')
plt.ylabel('Процентное изменение цен')

# 7. Добавляем легенду
plt.legend()

# 8. Отображаем график
plt.show()
```
